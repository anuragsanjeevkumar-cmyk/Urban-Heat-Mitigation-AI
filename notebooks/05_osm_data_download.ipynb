```python
import osmnx as ox

city = "Delhi, India"

buildings = ox.features_from_place(
    city,
    tags={"building": True}
)

roads = ox.features_from_place(
    city,
    tags={"highway": True}
)

print("Buildings:", buildings.shape)
print("Roads:", roads.shape)
```

```python
import osmnx as ox

city = "Delhi, India"

buildings = ox.features_from_place(
    city,
    tags={"building": True}
)

roads = ox.features_from_place(
    city,
    tags={"highway": True}
)

print("Buildings:", buildings.shape)
print("Roads:", roads.shape)
```

```python
buildings.head()
```

```python
roads.head()
```

```python
parks = ox.features_from_place(
    "Delhi, India",
    tags={"leisure": "park"}
)

water = ox.features_from_place(
    "Delhi, India",
    tags={"natural": "water"}
)

print("Parks:", parks.shape)
print("Water:", water.shape)
```

```python
import os

os.makedirs("../data/raw/osm", exist_ok=True)

buildings.to_file("../data/raw/osm/buildings.geojson", driver="GeoJSON")
roads.to_file("../data/raw/osm/roads.geojson", driver="GeoJSON")
parks.to_file("../data/raw/osm/parks.geojson", driver="GeoJSON")
water.to_file("../data/raw/osm/water.geojson", driver="GeoJSON")

print("All files saved successfully!")
```
