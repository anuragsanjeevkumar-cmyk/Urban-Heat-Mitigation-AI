# What is Urban Heat Island (UHI)?

Urban Heat Island (UHI) is a phenomenon where urban areas become significantly warmer than nearby rural areas.

Reasons:
- Concrete buildings absorb heat.
- Asphalt roads store heat.
- Lack of vegetation and trees.
- Vehicle and industrial emissions.
- High population density.

Effects:
- Increased energy consumption.
- Health risks during heatwaves.
- Poor air quality.
- Reduced comfort in cities.

# What is Land Surface Temperature (LST)?

Land Surface Temperature (LST) is the temperature of the Earth's surface measured by satellites.

Difference between Air Temperature and LST:

Air Temperature:
- Measured 2 meters above ground.
- Recorded by weather stations.

LST:
- Measured directly from the Earth's surface.
- Captured by satellite sensors.

Importance:
- Helps identify urban heat hotspots.
- Used for heat stress analysis.

# What is NDVI?

NDVI (Normalized Difference Vegetation Index) is used to measure vegetation health.

Formula:

NDVI = (NIR - Red) / (NIR + Red)

Interpretation:

NDVI > 0.6
Dense vegetation

NDVI 0.2 - 0.6
Moderate vegetation

NDVI < 0.2
Low vegetation

Relationship with Urban Heat:

More vegetation → Lower temperature

Less vegetation → Higher temperature

# What is Land Use Land Cover (LULC)?

LULC describes how land is being used and what covers the surface.

Examples:

- Forest
- Water Bodies
- Agricultural Land
- Residential Areas
- Industrial Areas
- Roads

Importance:

Different land cover types affect temperature differently.

Example:

Forest → Cooler

Concrete Buildings → Hotter
