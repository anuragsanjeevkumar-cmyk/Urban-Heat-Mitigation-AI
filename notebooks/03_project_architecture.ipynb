# Problem Understanding

Urban areas experience higher temperatures than surrounding rural regions due to dense infrastructure, reduced vegetation, and human activities.

The objective of this project is to use satellite imagery, weather data, and urban infrastructure data to identify heat hotspots, understand the causes of urban heating, predict land surface temperature, and recommend optimized cooling interventions.

# Inputs

Satellite Data:
- Landsat 8 LST
- ECOSTRESS LST
- Sentinel-2 LULC

Weather Data:
- ERA5
- CPCB

Urban Infrastructure:
- OpenStreetMap
- GHSL

# Outputs

- Heat Stress Maps
- Heat Hotspots
- Driver Analysis
- Cooling Recommendations
- Estimated Temperature Reduction

# Target Variable

Land Surface Temperature (LST)

The model will learn relationships between environmental factors and surface temperature.

# Scenario 1: Urban Greening

Increase vegetation cover.

Expected Impact:
Reduction in temperature.

# Scenario 2: Cool Roofs

Increase surface reflectivity.

Expected Impact:
Lower heat absorption.

# Scenario 3: Water Bodies

Introduce cooling through evaporation.

Expected Impact:
Reduction in urban heat stress.

# Scenario 4: Combined Strategy

Trees + Cool Roofs + Water Bodies

Expected Impact:
Maximum temperature reduction.
