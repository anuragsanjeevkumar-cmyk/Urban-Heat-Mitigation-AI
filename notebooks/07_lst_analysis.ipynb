```python
import rasterio
import matplotlib.pyplot as plt
import numpy as np

with rasterio.open("../data/raw/lst/Delhi_LST_2024.tif") as src:
    lst = src.read(1)

print("Shape:", lst.shape)
print("Min Temp:", np.nanmin(lst))
print("Max Temp:", np.nanmax(lst))

plt.figure(figsize=(10,8))
plt.imshow(lst, cmap="hot")
plt.colorbar(label="Temperature (°C)")
plt.title("Delhi Land Surface Temperature")
plt.show()
```

```python
import rasterio
import matplotlib.pyplot as plt

fig, ax = plt.subplots(1, 2, figsize=(15, 6))

with rasterio.open("../data/raw/ndvi/Delhi_NDVI_2024.tif") as src:
    ndvi = src.read(1)

with rasterio.open("../data/raw/lst/Delhi_LST_2024.tif") as src:
    lst = src.read(1)

ax[0].imshow(ndvi, cmap="RdYlGn")
ax[0].set_title("NDVI")

ax[1].imshow(lst, cmap="hot")
ax[1].set_title("LST")

plt.show()
```
